```python
import pandas as pd
import time

df = pd.read_csv('/content/StudentsPerformance.csv')

def stream_data(df):
    for i, row in df.iterrows():
        print("Streaming Record:", row.to_dict())
        time.sleep(0.5)

stream_data(df.head(10))
```

```python
def process_stream(record):
    avg = (record['math score'] + record['reading score'] + record['writing score']) / 3
    return avg

for _, row in df.head(5).iterrows():
    print("Processed Avg:", process_stream(row))
```
